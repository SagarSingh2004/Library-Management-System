# ðŸ“š Library Management System (LMS)

A simple console-based Library Management System built in Python.

```python
library2={
    "python_basics":{"Author":"author1","Availability":"available","Issued_to":"none"},
    "python_functions":{"Author":"author2","Availability":"available","Issued_to":"none"},
    "python_loops":{"Author":"author3","Availability":"available","Issued_to":"none"}
}
```

```python
print("Welcome to LMS portal")

credentials={"admin": "admin123", "librarian": "lib123"}
login_success=False


def dic_books():
    print("\n----Library books----")
    for keys,values in library2.items():
        Author=values["Author"]
        status=values["Availability"]
        issued_to=values["Issued_to"]
        if status=="available":
            print(f"\n{keys.title()} | Author : {Author.title()} | status : {status.title()}\n")
        else:
            print(f"\n{keys.title()} | Author : {Author.title()} | status : Issued to {issued_to.title()}\n")



def adding_books():
    print("\n----Adding a book----\n")
    title=input("Enter book name:").lower().strip()
    print()
    author=input("Enter the author of book:").lower().strip()
    print()
    if title in library2:
        if library2[title]["Author"]==author:
            print("\nThe book is already exists\n")
            dic_books()
            return
    library2[title]={"Author":author,"Availability":"available","Issued_to":"none"}
    print("\nSuccessfully added\n")
    dic_books()



def issuing_book():
    print("\n----Issuing a book----\n")
    user_input=input("Enter the book name to check its availability:").lower().strip()
    print()
    if user_input in library2:
        if library2[user_input]["Availability"]=="available":
            print("\nThe book is availble\n")
            issuer_name=input("Enter name of person issuing the book:").lower().strip()
            library2[user_input]["Availability"]="unavailable"
            library2[user_input]["Issued_to"]=issuer_name
            print(f"\nThe {user_input} is successfully issued to {issuer_name}\n")
        else:
            print(f"\nThe book is already issued to: {library2[user_input]["Issued_to"]}\n")
    else:
        print("\nThe book is not available\n")



def returning_book():
    print("\n----Returning a book----\n")
    user_input=input("Enter the book").lower().strip()
    print()
    if user_input in library2:
        if library2[user_input]["Availability"]=="unavailable":
            print(f"\nThe {user_input.title()} was issued to {library2[user_input]["Issued_to"]}\n")
            print(f"\n{user_input.title()} returned successfully by {library2[user_input]["Issued_to"]}.\n")
            library2[user_input]["Availability"]="available"
            library2[user_input]["Issued_to"]="none"
        else:
            print(f"\nThe {user_input} is not issued to anyone\n")
    else:
        print("\nThe book is not belongs to us\n")



for i in range(3):
    username=input("Enter the username").lower()
    if username in credentials:
        for j in range(3):
            password=input("Enter the password")
            if credentials[username]==password:
                print("\nSuccessfully login\n")
                login_success=True
                break
            else:
                print(f"\nEnter correct password, Attempt left: {2-j}\n")
        if login_success:
            break
        else:
            print("\nMax Attempt reached\n")
    else:
        print("\nIncorrect username Enter correct username\n")
        print(f"\n{2-i} Attempt left\n")
else:
    print("\nYou have exceeded maximum login attempts. Try again later.\n")



def view_issued_books():
    print("\nList of Issued Books:\n")
    Books_found=False
    for keys,values in library2.items():
        Author=values["Author"]
        status=values["Availability"]
        issued_to=values["Issued_to"]
        if status=="unavailable" and issued_to!="none":
            print(f"{keys.title()} | Issued to : {issued_to.title()}\n")
            Books_found=True
    if not Books_found:
        print("\nNo books are currently issued.\n")
    




def main_menu():
    while True:
        print("\n===========Options Available==========\n")
        print("\n1. View Books\n")
        print("\n2. Add a Book\n")
        print("\n3. Issue a Book\n")
        print("\n4. Return a book\n")
        print("\n5. View Issued Books\n")
        print("\n6. Exit\n")
    
        option=input("Enter your choice from (1 to 6)").strip()
    
        if option=="1":
            dic_books()
        elif option=="2":
            adding_books()
        elif option=="3":
            issuing_book()
        elif option=="4":
            returning_book()
        elif option=="5":
            view_issued_books()
        elif option=="6":
            print("\nExiting the Library, Goodbye!\n")
            break
        else:
            print("\nInvalid Choice\n")


if login_success==True:
    main_menu()
```
